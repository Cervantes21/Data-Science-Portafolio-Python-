## **Organización de los datos**

Primero crearemos una función para descargar nuestros datos de cada mes y así guardarlos en una lista, para que posteriormente los podamos guardar en variables, una vez que tengamos nuestras variables las uniremos a un solo conjunto de datos para poder realizar el **Análisis exploratorio de datos (EDA)**

<div style="background-color:#00798c;color:#edae49;font-family:monospace;text-align:center;font-size:16px;">
    
Creamos una nueva sesión para Pyspark

```python
# Configurar la sesión de Spark
spark = SparkSession.builder \
    .appName("Divvy Trip Data") \
    .getOrCreate()
```

```python
# Creamos una función para descargar nuestros datos de cada mes:

def download_file(url, filename):
    response = requests.get(url)
    with open(filename, 'wb') as f:
        f.write(response.content)
```

<div style="background-color:#edae49;color:#d1495b;font-family:monospace;padding-top:9px;font-size:16px;padding-left:7px;">
    
Crearemos una lista con las URL's de los archivos de cada mes para usarla en nuestro bucle para descargarlos y guardarlos con nombres descriptivos:

```python
urls = ['https://divvy-tripdata.s3.amazonaws.com/202201-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202202-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202203-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202204-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202205-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202206-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202207-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202208-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202209-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202210-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202211-divvy-tripdata.zip',
        'https://divvy-tripdata.s3.amazonaws.com/202212-divvy-tripdata.zip']

for url in urls:
  filename = url.split('/')[-1]
  download_file(url, filename)
```

```python
# Extraer el archivo CSV del archivo ZIP y cargar los datos en un DataFrame de Spark
dfs = []
for url in urls:
    filename = url.split('/')[-1]
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        csv_file = zip_ref.namelist()[0]
        zip_ref.extract(csv_file)
        df = spark.read.csv(csv_file, header=True, inferSchema=True)
        dfs.append(df)
```

<div style="background-color:#00798c;color:#edae49;font-family:monospace;text-align:center;font-size:16px;">
    
Unimos nuestros datos en un solo dataframe que contenga todos los meses.

```python
# Unir los DataFrames en uno solo
data = dfs[0]
for df in dfs[1:]:
    data = data.union(df)
```

### Imputación de los datos:
la mediana es una buena opción para imputar los valores faltantes en este caso. Ya que corresponden a los puntos de la latitud y la longitud de las estaciones.

```python
# Observamos nuestros datos nulos:
print(data_ps[['end_lat', 'end_lng']].isnull().sum())
```

```python
# Crear el objeto Imputer
imputer = Imputer(inputCols=['end_lat', 'end_lng'], outputCols=['end_lat_imputed', 'end_lng_imputed'])

# Ajustar el imputer al DataFrame
imputer_model = imputer.fit(data)

# Aplicar la imputación al DataFrame
data = imputer_model.transform(data)

# Sobreescribir las columnas originales con las columnas imputadas
data = data.withColumn('end_lat', col('end_lat_imputed')).withColumn('end_lng', col('end_lng_imputed'))

# Eliminar las columnas que no son necesarias
data = data.drop('end_lat_imputed', 'end_lng_imputed')
```

```python
# Verificar los valores nulos en las columnas 'end_lat' y 'end_lng'
null_counts = data.select([sum(col(c).isNull().cast("integer")).alias(c) for c in ['end_lat', 'end_lng']])

# Mostrar los resultados
null_counts.show()
```

Por lo tanto comenzaremos a arreglar nuestro conjunto de datos, transformado el tipo de formato de los datos para que podamos hacer un análisis más certero y podamos sacarle el máximo provecho al Dataset.

```python
# Convertir "started_at" y "ended_at" a objetos datetime
data = data.withColumn("started_at", to_timestamp(col("started_at")))
data = data.withColumn("ended_at", to_timestamp(col("ended_at")))

# Calcular la duración del viaje en minutos
data = data.withColumn('duration_min', (col('ended_at').cast('long') - col('started_at').cast('long')) / 60)

# Crear columnas para día de la semana y hora del día
data = data.withColumn("day_of_week", dayofweek(col("started_at")))
data = data.withColumn("hour_of_day", hour(col("started_at")))

# Obtener el año y mes de la columna "started_at"
data = data.withColumn("year", year(col("started_at")))
data = data.withColumn("month", month(col("started_at")))

# Convertir la columna "month" a tipo TIMESTAMP
data = data.withColumn("month", to_timestamp(col("month").cast("string"), "M"))

# Obtener el nombre del mes en español
data = data.withColumn("month", date_format(col("month"), "MMMM"))
```

```python
# comprobamos los cambios a nuestro conjunto:
data.printSchema()
```

<div style="background-color:#00798c;color:#edae49;font-family:monospace;font-size:16px;">
    
Ahora que hemos agregado las nuevas columnas, podemos hacer variables que contengan distintas agrupaciones, para que posteriormente sean convertidas en **Dataframes** y podamos analizarlos.

```python
# Agrupar por año, mes y tipo de miembro, y contar el número de viajes
trips_per_month_year_member = data.groupBy("year", "month", "member_casual").count()


# Agrupar los datos por "day_of_week" y "hour" y calcular estadísticas descriptivas de la duración del viaje
grouped_data = data.groupBy("day_of_week", "hour_of_day").agg(avg("duration_min").alias("avg_duration"),
                                                             max("duration_min").alias("max_duration"),
                                                             min("duration_min").alias("min_duration"))
# Reemplazar la columna "day_of_week" con los nombres de los días de la semana
grouped_data = grouped_data.withColumn("day_of_week",
                                       when(col("day_of_week") == 0, "Lunes")
                                       .when(col("day_of_week") == 1, "Martes")
                                       .when(col("day_of_week") == 2, "Miércoles")
                                       .when(col("day_of_week") == 3, "Jueves")
                                       .when(col("day_of_week") == 4, "Viernes")
                                       .when(col("day_of_week") == 5, "Sábado")
                                       .when(col("day_of_week") == 6, "Domingo"))


# Contar el número de viajes por hora del día y tipo de usuario
hour_counts = data.groupBy('hour_of_day', 'member_casual').agg(count('ride_id').alias('count')).select('hour_of_day', 'member_casual', 'count')

# Seleccionar las 10 estaciones más usadas por los usuarios en el año
top_stations = data.groupBy('start_station_name').count().orderBy(desc('count')).limit(10)

# Seleccionar las 10 estaciones más usadas donde finalizan los viajes los usuarios:
top_end_stations = data.groupBy('end_station_name').count().orderBy(desc('count')).limit(10)

# Filtrar los datos en dos grupos
annual_members = data.filter(data['member_casual'] == 'member')
casual_users = data.filter(data['member_casual'] == 'casual')


# Obtener las 10 estaciones más utilizadas por cada grupo
windowSpec = Window.partitionBy("start_station_name").orderBy(desc("count"))

top_stations_annual = annual_members.groupBy("start_station_name").agg(count("*").alias("count"))\
    .withColumn("rank", row_number().over(windowSpec))\
    .filter(col("rank") <= 10)

top_stations_casual = casual_users.groupBy("start_station_name").agg(count("*").alias("count"))\
    .withColumn("rank", row_number().over(windowSpec))\
    .filter(col("rank") <= 10)

# Obtener las 10 estaciones donde finalizan el viaje cada tipo de usuario:
windowSpec = Window.partitionBy("end_station_name").orderBy(desc("count"))

top_end_stations_annual = annual_members.groupBy("end_station_name").agg(count("*").alias("count"))\
    .withColumn("rank", row_number().over(windowSpec))\
    .filter(col("rank") <= 10)

top_end_stations_casual = casual_users.groupBy("end_station_name").agg(count("*").alias("count"))\
    .withColumn("rank", row_number().over(windowSpec))\
    .filter(col("rank") <= 10)

# Agrupar los datos por "member_casual" y calcular estadísticas descriptivas de la duración del viaje
ride_length = data.groupby('member_casual').agg(
    mean('duration_min').alias('mean'),
    stddev('duration_min').alias('stddev'),
    min('duration_min').alias('min'),
    max('duration_min').alias('max')
)

# Ordenar los resultados por "member_casual" de forma descendente
ride_length = ride_length.orderBy(desc('member_casual'))

# Agrupar por rideable_type y member_casual, y contar el número de viajes
rideable_type_data = data.groupBy("rideable_type", "member_casual").count().orderBy("rideable_type")
```

##### **TIPOS DE EVALUACIÓN:**

| Siglas | Descripción |
| -- | -- |
| VP | Verdadero Positivo |
| VN | Verdadero Negativo |
| FP | Falso Positivo |
| FN | Falso Negativo |

---

**Exactitud o accuracy**

- Proporciona cercanía al resultado de una medición del valor verdadero.

- En términos estadísticos, la exactitud está relacionada con el sesgo de una estimación.

- Proporción entre los positivos reales predichos y todos los casos postivos.

- En forma práctica, la exactitud es el % **total de elementos clasificados correctamente**

**Fórmula:**

**(VP+ VN) / (VP+FP+FN+VN) * 100**

---

**Precisión:**

- Dispersión del conjunto de valores obtenidos a partir de mediciones repetidas de una magnitud.

- Cuanto menor es la dispersión, mayor la precisión.

- Proporción entre el número de predicciones correctas (tanto positivas como negativas) y el total de predicciones.

- En forma práctica, es el **porcentaje de casos postitivos detectados**.

- Sirve para medir la **calidad del modelo** de machine learning en **clasificación**

**Fórmula:**

**(VP)/(VP+FP)**

---

**Sensibilidad:**

- Recall, sensitivity o **tasa de verdaderos positivos.**
- Proporción de **casos positivos** que fueron correctamente identificados.

**Fórmula de sensibilidad**

**VP/(VP+FN)**

---

**Especificidad:**
- **Tasa de verdaderos negativos**
- Proporción de **casos negativos** que fueron correctamente identificados

**Fórmula:**

**VN/VN+FP**

---

**F1-SCORE**
- Resume la **presición y sensibilidad** en una sola métrica

**Fórmula**

***2 * (precision * recall / precision + recall)***

---

#### **En resumen:**

- **Precision** = *TruePositive / Actual results or TruePositive / TruePositive + FalsePositive*

- **Recall** = *TruePositive / PredictResults or TruePositive / TruePositive + FalseNegative*

- **Accuracy** = *TruePositive + TrueNegative / Total*

- **Specificity** = *TrueNegative / TrueNegative + FalsePositives*

Podemos observar que nuestro modelo nuevamente sufre de sobre ajuste(**Overfitting**), ahora aplicaremos el uso de una técnica llamada **"búsqueda en cuadrícula"** (**GridSearchCV** en este caso) para realizar una búsqueda sistemática y exhaustiva en el espacio de hiperparámetros con el objetivo de encontrar la combinación que produce el mejor rendimiento del modelo.

Así podremos encontrar un modelo que se ajuste a predecir con mayor exactitud.

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Cambiar nuestras variables categóricas a numéricas
encoder = ce.OrdinalEncoder(cols=['member_casual','month'])
x_train = encoder.fit_transform(x_train)
x_test = encoder.transform(x_test)

# Definir el modelo de Random Forest
rf = RandomForestClassifier(random_state=0)

# Definir los hiperparámetros que deseamos probar
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 5, 10, 20]
}

# Realizar la búsqueda exhaustiva de hiperparámetros con validación cruzada
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

# Obtener los mejores hiperparámetros y el mejor modelo
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluar el modelo con los mejores hiperparámetros en el conjunto de prueba
y_test_pred_rf = best_model.predict(x_test)
test_accuracy_rf = accuracy_score(y_test, y_test_pred_rf)
print(f'Accuracy en test con los mejores hiperparámetros: {test_accuracy_rf:.2f}')

# Definir nombres de las clases
clases = ['classic_bike', 'docked_bike', 'electric_bike']

# Matriz de confusión
cm = confusion_matrix(y_test, y_test_pred_rf)

# Métricas calculadas
TP = np.diag(cm)
FP = np.sum(cm, axis=0) - TP
FN = np.sum(cm, axis=1) - TP

# Imprimir las métricas
for i, clase in enumerate(clases):
    print(f"Verdaderos positivos (TP) para \"{clase}\": {TP[i]:,}")
    print(f"Falsos positivos (FP) para \"{clase}\": {FP[i]:,}")
    print(f"Falsos negativos (FN) para \"{clase}\": {FN[i]:,}")
    print()
```

```python
# Graficar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=clases, yticklabels=clases)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Matriz de Confusión')
plt.savefig('./images/14_cm_rideable_type.jpg',bbox_inches='tight')
plt.show()

# Imprimir las métricas
for i, clase in enumerate(clases):
    print(f"Verdaderos positivos (TP) para \"{clase}\": {TP[i]:,}")
    print(f"Falsos positivos (FP) para \"{clase}\": {FP[i]:,}")
    print(f"Falsos negativos (FN) para \"{clase}\": {FN[i]:,}")
    print()

# La precisión, recall y f1-score para cada clase
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)

for i, clase in enumerate(clases):
    print(f"Precisión para \"{clase}\": {precision[i]:.2f}")
    print(f"Recall para \"{clase}\": {recall[i]:.2f}")
    print(f"F1-score para \"{clase}\": {f1_score[i]:.2f}")
    print()
```

### **Modelo de relación según mes y tipo de usuario en el año**.

Habra algún patrón que podamos observar en los datos sobre el mes y la relación al uso de los usuarios, para ello haremos una clusterización y utilizaremos el algoritmo **kmeans** para elaborar el modelo y posteriormente analizar los resultados.

```python
# Observemos los primeros valores de nuestro dataset:
trips_per_month_year_member_pd.head()
```

<div style="background-color:#00798c;color:#edae49;font-family:monospace;font-size:16px;">
    
En el tenemos los valores de:

- **year**: El año al que pertenece.
- **month**: El mes al que pertenece.
- **member_casual**: Nos da la categoría de los conteos si es un ciclista ocacional **casual** o un miembro anual **member**
- **count**: Nos proporciona los conteos de los usuarios para cada categoría según el mes y año.

```python
print(trips_per_month_year_member_pd.isnull().sum())
```

<div style="background-color:#edae49;color:#00798c;font-family:monospace;font-size:16px;">
    
Al buscar valores nulos en nuestro dataset, podemos ver que no tenemos valores nulos, por lo tanto podremos continuar a crear nuestro modelo de agrupación con **kmeans**.

```python
# Convertimos el tipo de dato de la columna "month" a numérico para poder utilizarlo en el algoritmo
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

trips_per_month_year_member_pd['month'] = trips_per_month_year_member_pd['month'].map(month_mapping)

# Seleccionamos las características que vamos a utilizar en el algoritmo
X = trips_per_month_year_member_pd[['month', 'count']]

# Aplicamos el algoritmo K-means con 2 clusters, uno para cada tipo de usuario
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
trips_per_month_year_member_pd['cluster'] = kmeans.fit_predict(X)

# Función para encontrar el mes más afluente para cada tipo de usuario
def find_most_frequent_month(df, cluster_label):
    most_frequent_month = df[df['cluster'] == cluster_label].groupby('month')['count'].sum().idxmax()
    return list(month_mapping.keys())[list(month_mapping.values()).index(most_frequent_month)]

# Función para graficar los clusters y los centroides usando Seaborn
def plot_clusters_seaborn(df):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x='month', y='count', hue='cluster', palette='Set1', s=100, alpha=0.7)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='x', s=200, label='Centroids')
    plt.xlabel('Month')
    plt.ylabel('Count')
    plt.title('Agrupación de usuarios por mes:')
    plt.legend()

    # Obtenemos las coordenadas del centroide para cada cluster
    centroids_x = kmeans.cluster_centers_[:, 0]
    centroids_y = kmeans.cluster_centers_[:, 1]

    # Agregamos el texto con la información de los meses más afluentes
    for i, centroid in enumerate(zip(centroids_x, centroids_y)):
        most_frequent_month = find_most_frequent_month(df, cluster_label=i)
        plt.text(centroid[0], centroid[1], f'Mes más afluente: {most_frequent_month}', ha='center', va='center', bbox=dict(facecolor='white', alpha=0.7))

    plt.savefig('./images/16_clustering_of_months_and_users.jpg', bbox_inches='tight')
    plt.show()


# Graficamos los clusters y los centroides usando Seaborn
plot_clusters_seaborn(trips_per_month_year_member_pd)

# Encontramos el mes más afluente para ciclistas ocasionales (Cluster 0)
most_frequent_casual_month = find_most_frequent_month(trips_per_month_year_member_pd, cluster_label=0)

# Encontramos el mes más afluente para miembros anuales (Cluster 1)
most_frequent_member_month = find_most_frequent_month(trips_per_month_year_member_pd, cluster_label=1)

print(f"El mes más afluente con más ciclistas ocasionales es: {most_frequent_casual_month}")
print(f"El mes más afluente com más miembros anuales es: {most_frequent_member_month}")
```

<div style="background-color:#00798c;color:#edae49;font-family:monospace;font-size:16px;">
    
Comprobamos que nuestro modelo puede predecir según el mes, el tipo de usuario que es.

Podemos observar que el mes con más **ciclistas ocacionales** registrado es **Julio** y con mayor número de registros para los **miembros anuales** es **Marzo**

### **Modelo de predicción según el día de la semana con más recorridos por cada tipo de usuario**:

<div style="background-color:#00798c;color:#edae49;font-family:monospace;font-size:16px;">
    
Para esto nuevamente vamos a realizar un **clustering** con respecto a la hora del día y la duración promedio del recorrido para poder predecir el día más afluente para cada tipo de usuario.

```python
grouped_data_pd.head()
```

```python
grouped_data_pd.info()
```

```python
# Cambiar nuestras variables categóricas a numéricas:
encoder = ce.OrdinalEncoder(cols=['day_of_week'])
grouped_data_pd = encoder.fit_transform(grouped_data_pd)

# Seleccionamos las características que vamos a utilizar en el algoritmo K-means
X = grouped_data_pd[['hour_of_day', 'avg_duration', 'max_duration', 'min_duration']]

# Aplicamos el algoritmo K-means con 2 clusters, uno para cada tipo de usuario
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
grouped_data_pd['cluster'] = kmeans.fit_predict(X)

# Diccionario para mapear valores numéricos de días de la semana a nombres
day_mapping = {
    0: 'Lunes',
    1: 'Martes',
    2: 'Miércoles',
    3: 'Jueves',
    4: 'Viernes',
    5: 'Sábado',
    6: 'Domingo'
    }

# Función para encontrar el día de la semana más afluente para cada tipo de usuario
def find_most_frequent_day(df, cluster_label):
    most_frequent_day_numeric = df[df['cluster'] == cluster_label]['day_of_week'].mode().iloc[0]
    return day_mapping[most_frequent_day_numeric]

# Encontramos el día de la semana más afluente para ciclistas ocasionales (Cluster 0)
most_frequent_casual_day = find_most_frequent_day(grouped_data_pd, cluster_label=0)

# Encontramos el día de la semana más afluente para miembros anuales (Cluster 1)
most_frequent_member_day = find_most_frequent_day(grouped_data_pd, cluster_label=1)

# Función para graficar los clusters y los centroides usando Seaborn
def plot_clusters_seaborn(df):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x='hour_of_day', y='avg_duration', hue='cluster', palette='Set1', s=100, alpha=0.7)

    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='x', s=200, label='Centroids')

    plt.xlabel('Hour of Day')
    plt.ylabel('Average Duration')
    plt.title('Agrupación de usuarios por hora del día y duración promedio:')
    plt.legend()

    # Mostrar la descripción debajo de la gráfica
    plt.text(0.5, -0.2, f"El día de la semana más afluente para ciclistas ocasionales es: {most_frequent_casual_day}", ha='center', va='center', transform=plt.gca().transAxes)
    plt.text(0.5, -0.25, f"El día de la semana más afluente para miembros anuales es: {most_frequent_member_day}", ha='center', va='center', transform=plt.gca().transAxes)
    plt.savefig('./images/17_clustering_of_hours_and_duration.jpg', bbox_inches='tight')
    plt.show()

# Graficamos los clusters y los centroides usando Seaborn
plot_clusters_seaborn(grouped_data_pd)
```
